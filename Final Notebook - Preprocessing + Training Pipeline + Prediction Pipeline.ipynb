```python
def get_topics(row):
    """
    Function to preprocess raw topic sentiments
    Input: raw dataframe row (see above)
    Processing: (<topic1> <sentiment1>,<topic2> <sentiment2>) --> ([topic1,topic2],[sentiment1,sentiment2])
    e.g., (garage service positive,ease of booking positive) --> ([garage service, ease of booking], [positive,positive])
    """
    topic_list = []
    sentiment_list = []
    for item in row:
        if type(item) ==str and 'positive' in item:
            topic_list.append(item.split('positive')[0].strip())
            sentiment_list.append('positive')
        elif type(item) ==str and 'negative' in item:
            topic_list.append(item.split('negative')[0].strip())
            sentiment_list.append('negative')
    return (topic_list,sentiment_list)
```

```python
raw_topics['topic_sentiment'] = raw_topics.apply(lambda row: get_topics(row),axis=1) 
```

```python
# redundant topics with very few samples to be replaced by more relevant topics 

replace_topic_dict = {'tyre agedot code' : 'tyre age/dot code',
                      'advisoragent service' : 'advisor/agent service',
                      'mobile fitter didn\'t arrive' : 'mobile fitter',
                      'refund timescale' : 'refund',
                      'refund not actioned' : 'refund',
                      'change of date' : 'change of date/time',
                      'change of time' : 'change of date/time',
                      'discount not applied' : 'discounts'}

def replace_redundant_topics(topic_sentiment_tuple):
    """
    Function to replace redundant topics and return a new modified topic sentiment tuple
    """
    topics = topic_sentiment_tuple[0]
    sentiments = topic_sentiment_tuple[1]
    
    # modify topic list according to replacement dict
    topics = [replace_topic_dict[topic] if topic in replace_topic_dict else topic for topic in topics]
    
    return (topics,sentiments)
```

```python
raw_topics['topic_sentiment'] = raw_topics['topic_sentiment'].apply(replace_redundant_topics) 
```

```python
# list of standard allowed punctuations
PUNCT=string.punctuation 
stop_words = list(set(stopwords.words('english')))
def is_allowed(s):
# checks if a character 's' is a valid character - alphanumeric, space or alphanumeric
# pre-cursor for the cleaning_func
    if s.isalnum():
        return True
    elif s==' ':
        return True
    elif s in PUNCT:
        return True
    return False

    
def cleaning_func(text):
    """
    This function performs cleaning operations on strings specifically for the use case of extracting paragraphs from files.
    It includes:
    Lower casing, tokenization, stop words removal, filtering to alphanumeric tokens only, and lemmatization.
    ***NOTE - Lemmatization is NOT performed using POS tagging.***
    RETURNS:
    ---------
    clean_text: str, text with minimal cleaning for downstream input to GloVE 
    lemm_text: str, lemmatized, lowercased text with stop words removed for input to topic modeling
    """
    
    clean_text = text.replace("\xa0"," ") # replace unicode by space to retain such paragraphs
    clean_text = re.sub('[.+-=_*!@#]+',' ',clean_text) # remove repeating characters '.+-=_*!@#'
    clean_text = clean_text.replace('[REDACTED]',' ') # remove reoccuring REDACTED tag from reviews 
    clean_text = re.sub(' +', ' ', clean_text) # remove repeating spaces
    clean_text = clean_text.strip()
    clean_text = ''.join(e for e in clean_text if is_allowed(e))
    
    word_tokens = word_tokenize(clean_text.lower()) 
    word_tokens = [w for w in word_tokens if not w in stop_words]
    alnum_tokens = [w for w in word_tokens if w.isalnum()]
    lemmas = [wordnet_lemmatizer.lemmatize(w) for w in alnum_tokens]
    
    lemm_text = " ".join(lemmas)
    return (clean_text, lemm_text)
```

### Spacy POS Tagging

```python
# Load English tokenizer, tagger, parser, NER and word vectors
nlp = spacy.load("en_core_web_lg")
```

```python
# custom cleaning operations for nouns and noun phrases

# extend stop_words with common adjectives seen in corpus 
# e.g., great service, excellent price, fast delivery etc
stop_words.extend(['easy', 'great','good','excellent','best','poor','fast','quick','brilliant','convenient'])

def remove_stopwords(texts):
    return [[word for word in word_tokenize(doc) if word not in stop_words] for doc in texts]

# def clean_word_list(texts):
#     new_texts = []
#     for text in texts:
#         if len(text):
#             new_texts.append(text)
#     return new_texts

def clean(clean_text):
    word_tokens = word_tokenize(clean_text.lower()) 
    word_tokens = [w for w in word_tokens if w.isalnum()]
    word_tokens = [wordnet_lemmatizer.lemmatize(w) for w in word_tokens]
    word_tokens = [w for w in word_tokens if w not in stop_words]

    lemm_text = " ".join(word_tokens)
    return lemm_text
```

```python
"""
For each review - find all relevant noun phrases
Process reviews to extract nouns and noun phrases
Remove stop words (extended set)
Remove phrases with more than 3 tokens
"""
np_list = []
noun_list = []
def spacy_process(text):
    doc = nlp(text.lower())

    # Analyze syntax
    noun_phrases = [chunk.text for chunk in doc.noun_chunks] 
    noun_phrases = [clean(text) for text in noun_phrases if clean(text)!='' ]
    np_list.append(noun_phrases)
    
    nouns = [token.lemma_ for token in doc if token.tag_ == "NN"]
    nouns = remove_stopwords(nouns)
    nouns = [text for text in nouns if len(text)>0]
    nouns = [noun[0] for noun in nouns]
    noun_list.append(nouns)
    
    all_phrases = list(set(nouns+noun_phrases))
    all_phrases = [p for p in all_phrases if len(p.split())<4]

    return all_phrases

processed_reviews['phrases'] = processed_reviews.clean_text.apply(spacy_process)
processed_reviews.head(2)
```

```python
def get_seed_words(reviews_df = processed_reviews, topic_categories = unique_topics):
    """
    Function to get seed phrases for each topic
    Inputs
    ------
    reviews_df - Dataframe with topic columns indicating sentiments
    topic_categories - list of topics

    Outputs
    -------
    topic_specs - dict with frequency of each phrase for specific topic reviews and its similarity score with topic text
                  e.g., topic_specs[topic] = [(<phrase1>,<phrase1_frequency>,<similarity_score>),(<phrase2>,...,...),...]
    seed_topic_dict - dict with top 20 words from each topic based on frequency and similarity
                      e.g., seed_topic_dict[topic] = [<list of seed phrases>]
    """

    topic_specs = {}
    seed_topic_dict = {}

    for topic in topic_categories:

        freq_similarity_list = []

        tmp = processed_reviews[processed_reviews[topic] != 0]['phrases'].values

        # count occurences in topic corpus for all phrases
        counter_obj = Counter(chain.from_iterable(tmp))

        freq = counter_obj.most_common()
        freq = [f for f in freq if f[1]>5] # eliminate phrases that have frequency <=5

        # calculate similarity for each phrase with actual topic text
        topic_doc = nlp(topic)
        for f in freq:
            freq_similarity_list.append((f[0],f[1],round(topic_doc.similarity(nlp(f[0])),3)))

        # take top 20 similar phrases
        freq_similarity_list.sort(key=itemgetter(2,1),reverse = True)
        topic_specs.update({topic:freq_similarity_list})
        topic_list = [f[0] for f in freq_similarity_list[:20]]

        # update seed topic dict if at least 1 seed phrase is found - eliminate topic otherwise
        if len(topic_list)>0:
            seed_topic_dict.update({topic:[f[0] for f in freq_similarity_list[:20]]})

    return topic_specs, seed_topic_dict

topic_specs, seed_topic_dict = get_seed_words()
```

#### Feature Generation Functions

```python
from flair.embeddings import DocumentPoolEmbeddings, WordEmbeddings
from flair.data import Sentence

# define the GloVe embedding specifications
pooled_embeddings = DocumentPoolEmbeddings([WordEmbeddings('glove')], pooling='mean')

def get_topic_dist(text, token_vectorizer = token_vectorizer, lda_model = glda_model):
    if type(text) == str:
        text= [text]
    X_ngrams = token_vectorizer.transform(text)
    topic_dist = lda_model.transform(X_ngrams)
    return topic_dist
    
def get_noun_phrase_topics(text):
    doc = nlp(text.lower())

    # get noun phrases
    np = [chunk.text for chunk in doc.noun_chunks] 
    np = [clean(text) for text in np if clean(text)!='' ]

    topic_dist_np = get_topic_dist(np)
    topic_dist_np = topic_dist_np.mean(axis=0)

    return topic_dist_np

def get_glove_vecs(text):
    
    text_= Sentence(text)
    pooled_embeddings.embed(text_)
    emb = text_.get_embedding().detach().numpy()
    return emb
```

#### Generate Features from Review text
1. Process reviews and find mean topic dist of all noun phrases
2. Review topic distributions
3. Get Review GloVe embeddings

```python
def get_review_features(row,review_features):
    r_id = row['review_id']
    processed_text = row['processed_text']
    clean_text = row['clean_text']
    
    topic_dist = get_topic_dist(processed_text)[0]
    topic_dist_np = get_noun_phrase_topics(clean_text)
    glove_emb = get_glove_vecs(clean_text)
    
    review_features.update({r_id:[topic_dist, topic_dist_np, glove_emb]})

review_features = {}
processed_reviews.apply(lambda row, review_features = review_features: get_review_features(row,review_features),axis=1)
```

#### Generate features from Topic Text
1. Topic Distribution of Topic Phrases
2. Glove vectors of Topic Phrases

```python
topic_features = {}
for topic in list(unique_topics):
    
    topic_dist = get_topic_dist(topic)[0]
    glove_emb = get_glove_vecs(topic)
    
    topic_features.update({topic:[topic_dist, glove_emb]})
```

```python
# http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
import itertools

def plot_confusion_matrix(cm, classes,
                          recall=False,
                          precision=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `recall=True`.
    """
    if recall:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Recall confusion matrix")
    elif precision:
        cm = cm.astype('float') / cm.sum(axis=0)[:, np.newaxis]
        print("Precision confusion matrix")
    else:
        print('Confusion matrix')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if recall or precision else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# generate features using review and topic feature generation functions above
feature_vec = test_data.apply(lambda row, tf = topic_features, rf = review_features:\
                                     np.concatenate([tf[row['Topic']][0],tf[row['Topic']][1], \
                                      rf[row['review_id']][0],rf[row['review_id']][1],rf[row['review_id']][2]]),axis=1)

feature_vec = np.vstack(feature_vec)

# scale the features using StandardScaler Object
feature_vec = scaler.transform(feature_vec) 

# fix issue with numeric overflow
feature_vec = np.float32(feature_vec)
feature_vec = np.nan_to_num(feature_vec, nan=0.0, posinf=None, neginf=None)

# check dimensions of feature_vec
assert feature_vec.shape[1] == 272, "feature space doesn't have correct dimensions"
assert feature_vec.shape[0] == len(test_data), "feature space doesn't have correct dimensions"
```
